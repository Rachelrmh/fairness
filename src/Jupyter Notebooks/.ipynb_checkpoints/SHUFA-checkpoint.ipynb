```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from datetime import datetime
```

```python
#load SCHUFA data as csv
SCHUFA_DATA_DIR = "C:\Machine_Learning\ML Projects\Fairness\\fairness\data\statlog_german.data.txt"
raw_data = pd.read_csv(SCHUFA_DATA_DIR, sep=" ", header=None)
raw_data.head()
data = np.array(raw_data) #convert data to numpy matrix
print(data.shape)
raw_data.head()
```

```python

#Preprocessing
def label_encode(data, categories=None):
    '''
    Encode given categories(columns) using label encoding.
    
    Parameters:
        data: Data of type numpy.ndarray
        categories: A list(or tuple) of the columns/categories to label encode
    
    Returns:
        edited_data: A numpy matrix similar to data with label encoding on specified columns
    '''
    edited_data = None
    categories = set(categories)
    le = LabelEncoder()
    
    #Label encode each specified column
    for column in range(data.shape[1]):
        if column in categories:
            
            encoded_column = le.fit_transform(data[:, column])
            
            if type(edited_data) == type(None): 
                edited_data = np.reshape(encoded_column, (encoded_column.shape[0], -1))
            else:
                edited_data = np.concatenate((edited_data, np.reshape(encoded_column, (edited_data.shape[0], -1))), axis=1)
            
        else:
            if type(edited_data) == type(None):
                edited_data = [data[:, column]]
            else:   
                edited_data = np.concatenate((edited_data, np.reshape(data[:, column], (edited_data.shape[0], -1))), axis=1)
    return(edited_data)

def one_hot_encode(data, categories=None):
    '''
    Encode given categories(columns) using label encoding.
    
    Parameters:
        data: Data of type numpy.ndarray
        categories: A list(or tuple) of the columns/categories to label encode
    
    Returns:
        edited_data: A numpy matrix containing data with specified columns one hot encoded
    '''
    edited_data = None
    categories = set(categories)
    ohe = OneHotEncoder()
    for column in range(data.shape[1]):
        if column in categories:
            
            encoded_column = ohe.fit_transform(np.reshape(data[:, column], [data.shape[0], 1])).todense()
           
            if type(edited_data) == type(None):
                print(encoded_column.shape)
                edited_data = np.reshape(encoded_column, (encoded_column.shape[0], -1))
            else:
                edited_data = np.concatenate((edited_data, np.reshape(encoded_column, (edited_data.shape[0], -1))), axis=1)
        else:
            
            if type(edited_data) == type(None):
                edited_data = [data[:, column]]
            else:   
                edited_data = np.concatenate((edited_data, np.reshape(data[:, column], (edited_data.shape[0], -1))), axis=1)
    return(edited_data)

def mean_normalize_columns(data, columns):
    '''
    Normalize specified columns with mean normalization.
    
    Parameters: 
        data: Data of type numpy.ndarray
        columns: A list(or tuple) of the columns/categories to label encode
    
    Returns:
        edited_data: A numpy matrix the same shape as data with specified columns normalized
    '''
    edited_data = np.copy(data)
    for column in columns:
        column_mean = np.mean(data[:, column]) 
        column_std = np.std(data[:, column])
        edited_data[:, column] = np.subtract(data[:, column], column_mean)
        edited_data[:, column] = np.divide(edited_data[:, column], column_std)
        print(column, "Mean:", column_mean, "STD:", column_std, "Min Val:", np.amin(data[:, column]), "Max Val:", np.amax(data[:, column]))
        
    return(edited_data)

#Normalize columns 1, 4, 12
normalized_data = mean_normalize_columns(data, [1, 4, 12])

#Label encode and one hot encode columns 0, 2, 3, 5, 6, 8, 9, 11, 13, 14, 16, 18, 19
proccessed_data = label_encode(normalized_data, [0, 2, 3, 5, 6, 8, 9, 11, 13, 14, 16, 18, 19])
input_data = one_hot_encode(proccessed_data, [0, 2, 3, 5, 6, 8, 9, 11, 13, 14, 16, 18, 19])

#split data and labels
[input_data, labels] = np.split(input_data, [-1], axis=1)

#Since labels are orginally 1 for bad and 2 for good, subtract 1 to get 0 for bad and 1 for good
labels = np.reshape(np.subtract(labels, [1]), (-1, 1))
print(input_data.shape, labels.shape)

```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(input_data, labels):
    train_data, train_labels = input_data[train_index], labels[train_index]
    test_data, test_labels = input_data[test_index], labels[test_index]

print("Train Data Shape:", train_data.shape, "Train Labels Shape:", train_labels.shape)
print("Test Data Shape:", test_data.shape, "Test Labels Shape:", test_labels.shape)
```

```python
#Helper functions to create Tensorflow Deep Neural Network Graph

def weights(shape):
        '''
        Creates the filters and bias' for the convolutions/pooling layers of the inception_module

        Parameters:
            shape: The shape of the weight tensor.
                    For dense_layers, this should be [num_inputs_to_layer, num_outputs_of_layer].
                    Note: Number of outputs for a layer is equivalent to the number of neurons in that layer.
        Returns:
            The tensor weight variable
        '''
        return(tf.Variable(tf.truncated_normal(shape, stddev=0.1, name="weights")))

def bias(shape, initial_value=0.1):
    '''
    Create a bias variable. Note it is initalized by default to 0.1
        
    Parameters:
        shape: The shape of the bias tensor. This should be a 1D shape equal to the number of outputs of
                a layer.
        initial_value: The initial value of all the bias's
            
    Returns:
           The tensor bias variable 
    '''
    return(tf.Variable(tf.constant(initial_value, shape=shape, name="bias")))

def dense_layer(inputs, weights, bias, activation=None):
    '''
    Build a fully connected dense neuron layer. 
    
    Parameters:
        inputs: An input tensor with shape [batch_size, num_inputs]
        weights: A tensor with shape [num_inputs, num_outputs]
        bias: A tensor with shape [num_outputs]
        activation: A valid tensorflow activation function.(EX. tf.nn.relu())
    Returns:
        If activation is none, return the output of Wx+b. Else if activation is set, return
        the activation function with input Wx+b.
    '''
    output = tf.matmul(inputs, weights) + bias
    if activation is None:
        return(output)
    return(activation(output))


```

```python
def projection(weights, p_norm, lamda):
    '''
    Projection function on a weights matrix to constrain the lipschitz constant.
    
    Parameters:
        weights: The weights tenosr of a neuron layer
        p_norm: The The distance norm to be computed for the weights
        lamda: Hyperparameter to control the upper bound of the Lipschitz constant
    
    Returns: 
        Assignment operation to update the weights tensor.
        weights_norm: The p_norm of the weights tensro
    '''
    one_scalar = tf.constant(1, dtype=tf.float32)
    lamda = tf.constant(lamda, dtype=tf.float32)
    
    #compute the p_norm of the weight matrix
    weights_norm = tf.norm(weights, ord=p_norm)
    weight_bound = tf.divide(weights_norm, lamda)
    projection_parameter = tf.divide(one_scalar, tf.maximum(one_scalar, weight_bound))
    weights_projection = tf.multiply(projection_parameter, weights)
    
    return(tf.assign(weights, weights_projection), weights_norm)
    
```

```python
#Network model
num_inputs = 61
num_hidden1 = 20
num_hidden2 = 20
num_hidden3 = 1

with tf.name_scope("dnn_model"):
    
    #insertion point for input data to model
    X = tf.placeholder(dtype=tf.float32, shape=[None, num_inputs])

    #hidden layer 1
    w1 = weights([num_inputs, num_hidden1])
    b1 = bias([num_hidden1])
    z1 = dense_layer(X, w1, b1, activation=tf.nn.relu)

    #hidden layer 2
    w2 = weights([num_hidden1, num_hidden2])
    b2 = bias([num_hidden2])
    z2 = dense_layer(z1, w2, b2, activation=tf.nn.relu)

    #hidden layer 3
    w3 = weights([num_hidden2, num_hidden3])
    b3 = bias([num_hidden3])
    z3 = dense_layer(z2, w3, b3, activation=None)
    
    #Sigmoid function outputs probability between 0 and 1
    logits = tf.nn.sigmoid(z3)
```

# Logistic Regression Cost Function
Logistic Regression using a sigmoid function outputs a number(probability) between 0 and 1. With respect to this model, the probability represents how good a customers credit is. 

*Logistic Function*
$$\sigma (t)=\frac{1}{1+e^{-t}}$$

To make predicition for binary classifiction, the *Logistic Regression Model Predicition* is given by
$$prediction = \left\{\begin{matrix}
0 \quad if\ \sigma(t) < 0.5\\ 
1 \quad if\ \sigma(t) \geq 0.5
\end{matrix}\right.$$

The *Logistic Regression Cost Function(log loss)* penalizes incorrect predicitions.
$$J(\Theta ) = \frac{-1}{m}\sum_{i=1}^{m}\left [ y^{(i)}log(p^{(i)}) + (1-y^{(i)})log(1-p^{(i)}) \right ]$$
where $y^{(i)}$ is the label for instance $i$ and $p^{(i)}$ is the probability(logit) of instance $i$


```python
#Loss(cost) funtion
with tf.name_scope("logistic_cost"):
    
    y = tf.placeholder(dtype=tf.float32, shape=[None, 1])
    cost = tf.losses.log_loss(labels=y, predictions=logits)
    
```

# Training Deep Neural Network with Gradient Descent
The gradient of the cost function with respect to each weight $\theta$ is given by the *Logistic Cost Function Partial Derivatives* 
$$\frac{\partial }{\partial \theta_{j}}=\frac{1}{m}\sum_{i=1}^{m}(\sigma(\theta^{T}\cdot x^{(i)})-y^{(i)})x_{j}^{(i)}$$

After a forward pass through the network with the training mini-batches, backpropogation is used to update these weights in order to minimize the cost. 

```python
#Training Operation
learning_rate = 0.001
with tf.name_scope("training_op"):
    
    #Use gradient descent to reduce loss
    #optimizer = tf.train.GradientDescentOptimizer(learning_rate)
    optimizer = tf.train.MomentumOptimizer(learning_rate, momentum=.9, use_nesterov=True)
    #computes gradients and applys gradients
    training_op = optimizer.minimize(cost)
```

```python
prediction_thresh = tf.constant(0.5, dtype=tf.float32)
predictions=tf.cast(tf.greater_equal(logits, prediction_thresh), tf.float32)

with tf.name_scope("train_accuracy"):
    train_acc, train_acc_op = tf.metrics.accuracy(labels=y, predictions=predictions, name="train_acc")
    
    #isolate running variable for accuracy metric
    train_acc_running_vars = tf.get_collection(tf.GraphKeys.LOCAL_VARIABLES, scope="train_acc")
    train_acc_running_vars_initializer = tf.variables_initializer(var_list=train_acc_running_vars)
    
with tf.name_scope("test_accuracy"):
    test_acc, test_acc_op = tf.metrics.accuracy(labels=y, predictions=predictions, name="test_acc")
    
    #isolate running variable for accuracy metric
    test_acc_running_vars = tf.get_collection(tf.GraphKeys.LOCAL_VARIABLES, scope="test_acc")
    test_acc_running_vars_initializer = tf.variables_initializer(var_list=test_acc_running_vars)
    
```

```python
with tf.name_scope("train_precision"):
    train_prec, train_prec_op = tf.metrics.precision(labels=y, predictions=predictions, name="train_prec")
    
    #isolat running local variables for precision metric
    train_prec_running_vars = tf.get_collection(tf.GraphKeys.LOCAL_VARIABLES, scope="train_prec")
    train_prec_running_vars_initializer = tf.variables_initializer(var_list=train_prec_running_vars)
    
with tf.name_scope("test_precision"):
    test_prec, test_prec_op = tf.metrics.precision(labels=y, predictions=predictions, name="test_prec")
    
    #isolat running local variables for precision metric
    test_prec_running_vars = tf.get_collection(tf.GraphKeys.LOCAL_VARIABLES, scope="test_prec")
    test_prec_running_vars_initializer = tf.variables_initializer(var_list=test_prec_running_vars)

with tf.name_scope("train_recall"):
    train_rec, train_rec_op = tf.metrics.recall(labels=y, predictions=predictions, name="train_rec")
    
    #isolat running local variables for precision metric
    train_rec_running_vars = tf.get_collection(tf.GraphKeys.LOCAL_VARIABLES, scope="train_rec")
    train_rec_running_vars_initializer = tf.variables_initializer(var_list=train_rec_running_vars)
    
with tf.name_scope("test_recall"):
    test_rec, test_rec_op = tf.metrics.recall(labels=y, predictions=predictions, name="test_rec")
    
    #isolat running local variables for precision metric
    test_rec_running_vars = tf.get_collection(tf.GraphKeys.LOCAL_VARIABLES, scope="test_rec")
    test_rec_running_vars_initializer = tf.variables_initializer(var_list=test_rec_running_vars)
    
```

```python
with tf.name_scope("lipshitz_regularization"):
    p_norm=1
    lamda1 = 90
    lamda2 = 25
    lamda3 = 1
    #layer one regularization
    w1_projection, w1_norm = projection(w1, p_norm=p_norm, lamda=lamda1)
    w2_projection, w2_norm = projection(w2, p_norm=p_norm, lamda=lamda2)
    w3_projection, w3_norm = projection(w3, p_norm=p_norm, lamda=lamda3)
```

```python
#Helper function to retreive mini-batchs
def get_mini_batch(X, y, current_batch_num, batch_numbers):
    '''
    Get mini-batch of input data and labels. Each dataset will be split in n number of batches, and 
    current batch number will select the necessary mini-batch from the subsets of the split.
    
    Parameters:
        X: Input dataset to be fed for training into the DNN
        y: The target dataset for the input data used to train the DNN on
        current_batch_num: The current iteration in training to select a new mini-batch.
        batch_numbers: The overall number of batchs to split the whole dataset into
    
    Returns:
        batch_X: The mini-batch for input training data
        batch_y: The mini-batch for target training data
    '''
    X, y= np.asarray(X), np.asarray(y)
    batch_X = np.split(X, batch_numbers)[current_batch_num]
    batch_y = np.split(y, batch_numbers)[current_batch_num]
    return batch_X, batch_y
```

```python
epochs = 500
batch_size = 50

global_variable_init = tf.global_variables_initializer()
tf.get_default_graph()

num_batches = train_data.shape[0]//batch_size

#Evaluation metrics(for plotting)
train_cost = []
test_cost = []
test_acc_score = []
train_acc_score = []
test_prec_score = []
train_prec_score = []
test_rec_score = []
train_rec_score = []
w1_norm_data = []
w2_norm_data = []
w3_norm_data = []

with tf.Session() as sess:
    
    #Initialize global tensorflow variables
    global_variable_init.run()
    
    #Initialize local accuracy metrics variables
    sess.run([train_acc_running_vars_initializer, test_acc_running_vars_initializer])
    #Intializer local precision metrics variables
    sess.run([train_prec_running_vars_initializer, test_prec_running_vars_initializer])
    #Initialize local recall metrics variables
    sess.run([train_rec_running_vars_initializer, test_rec_running_vars_initializer])
    
    #Mini-batch gradient descent training
    for epoch in range(epochs):
        
        #Reset local updating variables for training evaluation metrics. Re-evaluate metrics each epoch
        sess.run(train_acc_running_vars_initializer)
        sess.run(train_prec_running_vars_initializer)
        sess.run(train_rec_running_vars_initializer)
        
        for batch in range(num_batches):
            #get mini-batch
            train_batch_data, train_batch_labels = get_mini_batch(train_data, train_labels, batch, num_batches)
            
            #Perform training operation by using mini-batch gradient descent
            sess.run(training_op, feed_dict={X:train_batch_data, y:train_batch_labels})
            
            #run training evaluation metrics
            sess.run([train_acc_op, train_prec_op, train_rec_op], feed_dict={X:train_batch_data, y:train_batch_labels})
            
            #Perform Lipshcitz constraint on weights
            #sess.run([w1_projection, w2_projection, w3_projection])
        
        #Save the norms of the weight matrices
        w1_norm_data.append(sess.run(w1_norm))
        w2_norm_data.append(sess.run(w2_norm))
        w3_norm_data.append(sess.run(w3_norm))
        
        #compute cost for test and training data
        test_cost.append(sess.run(cost, feed_dict={X:test_data, y:test_labels}))
        train_cost.append(sess.run(cost, feed_dict={X:train_batch_data, y:train_batch_labels}))
        
        #compute accuracy, precision, and recall for test data
        #reset running test accuracy, precision, and recall variables to 0
        sess.run(test_acc_running_vars_initializer)
        test_acc_score.append(sess.run(test_acc_op, feed_dict={X:test_data, y:test_labels}))
        sess.run(test_prec_running_vars_initializer)
        test_prec_score.append(sess.run(test_prec_op, feed_dict={X:test_data, y:test_labels}))
        sess.run(test_rec_running_vars_initializer)
        test_rec_score.append(sess.run(test_rec_op, feed_dict={X:test_data, y:test_labels}))
        
        #append training accuracy, recall, and precision scores
        train_acc_score.append(sess.run(train_acc))
        train_prec_score.append(sess.run(train_prec))
        train_rec_score.append(sess.run(train_rec))
        
        print("Training Progress {:2.1%}".format(float((epoch+1)/epochs)), end="\r", flush=True)
        
```

```python
min_test_loss = min(test_cost)
#Plot cost data
plt.figure(1)
plt.grid(True)
plt.title("Log. Cost")
plt.xlabel("Epochs")
plt.ylabel("Cost")
plt.plot(test_cost, label="Test")
plt.plot(train_cost, label="Train")
plt.hlines(min_test_loss, xmin=0, xmax=epochs, colors="r", linestyle="dashed", label="Min. Test Cost="+str(min_test_loss))
plt.legend()

#Plot accuracy data
plt.figure(2)
plt.grid(True)
plt.title("Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.plot(test_acc_score, label="Test")
plt.plot(train_acc_score, label="Train")
plt.legend()

#Plot accuracy data
plt.figure(3)
plt.grid(True)
plt.title("Precision/Recall")
plt.xlabel("Epochs")
plt.ylabel("Precision/Recall")
plt.plot(test_prec_score, label="Test_Prec.")
plt.plot(train_prec_score, label="Train_Prec.")
plt.plot(test_rec_score, label="Test_Rec.")
plt.plot(train_rec_score, label="Train_Rec.")
plt.legend()

#plot weight norms data
plt.figure(4)
plt.grid(True)
plt.xlabel("Epochs")
plt.ylabel("l"+str(p_norm)+" norm")
plt.plot(w1_norm_data, label="w1_norm")
plt.hlines(lamda1, xmin=0, xmax=epochs, colors="m", linestyle='dashed', linewidth=4, label="Lamda1="+str(lamda1))
plt.plot(w2_norm_data, label="w2_norm")
plt.hlines(lamda2, xmin=0, xmax=epochs, colors="k", linestyle='dashed', linewidth=4, label="Lamda2="+str(lamda2))
plt.plot(w3_norm_data, label="w3_norm")
plt.hlines(lamda3, xmin=0, xmax=epochs, colors="r", linestyle='dashed', linewidth=4, label="Lamda3="+str(lamda3))
plt.legend()

plt.show()

```
